```python
import matplotlib.pyplot as plt
import dayplot as dp
import pandas as pd

df = dp.load_dataset()  # requires pandas to be installed

fig, ax = plt.subplots(figsize=(15, 6))
dp.calendar(
    dates=df["dates"],
    values=df["values"],
    start_date="2024-01-01",
    end_date="2024-12-31",
    ax=ax,
)
```
